```python
import json
import os
```

```python
json_directory = "./202310"

for i, file_name in enumerate(os.listdir(json_directory)):
    if  i < 10 and file_name.endswith(".json"):
        file_path = os.path.join(json_directory, file_name)
        with open(file_path, "r") as file:
            data = json.load(file)
            print(f"Contents of {file_name}:")
            if data["submissions"] is not None and isinstance(data["submissions"], list):
                for j, submission in enumerate(data["submissions"]):
                    print(f"Submission: {j + 1}")
                    # print(content)
                    # print(content["iScore"])
                    if submission["iScore"] == "0":
                        print("Wrong Submission: ")
                    else:
                        print("Correct Submission: ")

                    print(submission["sourceCode"]["sSource"])
                    print()
                print()
```
